```python
#reading csv
raw_census = pd.read_csv('/content/drive/Shareddrives/Rice Datathon 2023/Copy of sc-est2019-agesex-civ.csv')
raw_census = raw_census[raw_census['NAME']!='United States']
raw_census = raw_census[(raw_census['SEX'] == 2) & (raw_census['AGE'] >= 50)]
raw_census = raw_census[['NAME', 'POPEST2010_CIV', 'POPEST2011_CIV', 'POPEST2012_CIV', 'POPEST2013_CIV', 'POPEST2014_CIV',
                          'POPEST2015_CIV', 'POPEST2016_CIV', 'POPEST2017_CIV', 'POPEST2018_CIV', 'POPEST2019_CIV']]
                          
raw_census.head()
```

```python
# Iterate through each column in the DataFrame (excluding the first column)
for col in raw_census.columns[1:]:
    # Extract characters 7 through 10 of the column name
    year = col[6:10]
    # Rename the column
    raw_census = raw_census.rename(columns={col: year})
    
raw_census
```

```python
# create a new dataframe with the state name and a column for each year
raw_census = raw_census.groupby('NAME').sum().reset_index()

# Print the DataFrame to confirm the new column has been added
raw_census
```

```python
# Load your dataframe
df = raw_census

# Create a new column that calculates the population change over the years
# df['population_change'] = df.iloc[:,1:].sum(axis=1) - df.iloc[:,1:].sum(axis=1).min()

# df.head()


# # Sort the states by population change in descending order
# df = df.sort_values(by='population_change', ascending=False)

# # Select the top n states with the most population change
# n = 10 # or any number of states you want to consider
# df_top_states = df.head(n)

# Create a new dataframe with the population estimates for the top states

if 'Rate' not in df:
  df.insert(11, "Rate", 0)

```

```python
# Create a new dataframe with the state names


# Create a linear regression object
years = np.array([2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019])

i = 0

# Fit the model using the population estimates
for row in df.iterrows():
  reg = LinearRegression().set_params(fit_intercept=True)
  #curr = np.array(row[1:][0][1:] / row[1:][0][1]) # Divide each by 2010
  curr = np.array(row[1:][0][1:11])
  #print(curr)
  reg = reg.fit(years.reshape(-1, 1), curr)

  # Print the coefficients of the linear regression
  
  #print(reg.coef_, reg.intercept_)

  df.iloc[i, 11] = (reg.coef_*2024 + reg.intercept_ - row[1:][0][-2]) / row[1:][0][-2]
  i += 1
df = df.sort_values(by=['Rate'], ascending=False)

df
```

```python
# Change the state name from complete to abbreviation
df["NAME"] = df["NAME"].replace({"Alabama":"AL", "Alaska":"AK", "Arizona":"AZ", "Arkansas":"AR", "California":"CA",
                                  "Colorado":"CO", "Connecticut":"CT", "Delaware":"DE", "District of Columbia":"DC", "Florida":"FL", "Georgia":"GA",
                                  "Hawaii":"HI", "Idaho":"ID", "Illinois":"IL", "Indiana":"IN", "Iowa":"IA", "Kansas":"KS",
                                  "Kentucky":"KY", "Louisiana":"LA", "Maine":"ME", "Maryland":"MD", "Massachusetts":"MA",
                                  "Michigan":"MI", "Minnesota":"MN", "Mississippi":"MS", "Missouri":"MO", "Montana":"MT",
                                  "Nebraska":"NE", "Nevada":"NV", "New Hampshire":"NH", "New Jersey":"NJ", "New Mexico":"NM",
                                  "New York":"NY", "North Carolina":"NC", "North Dakota":"ND", "Ohio":"OH", "Oklahoma":"OK",
                                  "Oregon":"OR", "Pennsylvania":"PA", "Rhode Island":"RI", "South Carolina":"SC", 
                                  "South Dakota":"SD", "Tennessee":"TN", "Texas":"TX", "Utah":"UT", "Vermont":"VT",
                                  "Virginia":"VA", "Washington":"WA", "West Virginia":"WV", "Wisconsin":"WI", "Wyoming":"WY"})
pop_plus_fac = df.merge(facilities_per_state, how="inner", left_on="NAME", right_on="State")
pop_plus_fac
```

```python
pop_plus_fac['Pop/Fac'] = pop_plus_fac['2019'] / pop_plus_fac['City']
pop_per_facility = pop_plus_fac[['State','Pop/Fac']]
#pop_per_facility
print(pop_per_facility['Pop/Fac'].max())
print(pop_per_facility['Pop/Fac'].min())
pop_per_facility.to_csv('/content/drive/Shareddrives/Rice Datathon 2023/women_per_facility.csv', index=False)
```
