```python
#standard imports
import pandas as pd
import numpy as np
import plotly.express as px
from google.colab import drive
```

```python
#reading csv
#note that this is 2019
#note that this age-adjusted means adjusted for w
raw_cancer = pd.read_csv('/content/drive/Shareddrives/Rice Datathon 2023/USCSOverviewMap (3).csv')
```

```python
cancer_cleaned = raw_cancer[['Area', 'Age-Adjusted Rate', 'Death Count', 'Population']]
```

```python
#cancer_cleaned.insert(3, "Raw Rate", cancer_cleaned['Death Count'] / cancer_cleaned['Population'])
#cancer_cleaned = cancer_cleaned[['Area', 'Death Count', 'Raw Rate', 'Population']]
cancer_cleaned['Raw Rate'] = cancer_cleaned['Death Count'] / cancer_cleaned['Population'] * 100000 # Deaths per 100,000
cancer_cleaned = cancer_cleaned.sort_values('Raw Rate')
cancer_cleaned = cancer_cleaned.rename(columns = {'Raw Rate':'Raw Mortality Rate'})

cancer_cleaned = cancer_cleaned[cancer_cleaned['Area']!='Puerto Rico']
```

```python
cancer_cleaned["Area"] = cancer_cleaned["Area"].replace({"Alabama":"AL", "Alaska":"AK", "Arizona":"AZ", "Arkansas":"AR", "California":"CA",
                                  "Colorado":"CO", "Connecticut":"CT", "Delaware":"DE", "District of Columbia":"DC", "Florida":"FL", "Georgia":"GA",
                                  "Hawaii":"HI", "Idaho":"ID", "Illinois":"IL", "Indiana":"IN", "Iowa":"IA", "Kansas":"KS",
                                  "Kentucky":"KY", "Louisiana":"LA", "Maine":"ME", "Maryland":"MD", "Massachusetts":"MA",
                                  "Michigan":"MI", "Minnesota":"MN", "Mississippi":"MS", "Missouri":"MO", "Montana":"MT",
                                  "Nebraska":"NE", "Nevada":"NV", "New Hampshire":"NH", "New Jersey":"NJ", "New Mexico":"NM",
                                  "New York":"NY", "North Carolina":"NC", "North Dakota":"ND", "Ohio":"OH", "Oklahoma":"OK",
                                  "Oregon":"OR", "Pennsylvania":"PA", "Rhode Island":"RI", "South Carolina":"SC", 
                                  "South Dakota":"SD", "Tennessee":"TN", "Texas":"TX", "Utah":"UT", "Vermont":"VT",
                                  "Virginia":"VA", "Washington":"WA", "West Virginia":"WV", "Wisconsin":"WI", "Wyoming":"WY"})
if "MR per State" not in cancer_cleaned:
  cancer_cleaned.insert(5, 'MR per State', 0)
cancer_cleaned = cancer_cleaned.sort_values(by=['Area'])
cancer_cleaned = cancer_cleaned.reset_index()
cancer_cleaned
```

```python
from numpy.lib.function_base import delete
num_per_state = pd.read_csv('/content/drive/Shareddrives/Rice Datathon 2023/facilities_per_state.csv')
num_per_state.drop(columns="Unnamed: 0")
num_per_state = num_per_state.sort_values(by="State")
fin = cancer_cleaned
states = ["AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DC", "DE", "FL", "GA", 
          "HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD", 
          "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ", 
          "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC", 
          "SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY"]
states.sort()
i = 0

for state in states:
  fin.loc[i, 'MR per State'] = fin.loc[i, 'Raw Mortality Rate'] / num_per_state.loc[i, 'City']
  i += 1

sorted_fin = fin.sort_values(by="MR per State")
sorted_fin
```

```python
sorted_fin.to_csv('/content/drive/Shareddrives/Rice Datathon 2023/cancer_per_state.csv', index=False)
```
